```python
import os
import csv

#import csv file
election_csv = os.path.join('/Users/ahillike/Desktop/git_projects/python-challenge-hilliker/PyPoll/election_data.csv')
with open(election_csv, newline = "") as csvfile:
    csvreader = csv.reader(csvfile, delimiter = ",")
    
    #Read the header row first and go to next row
    csv_header = next(csvreader)
    #print(csv_header)
    
    #Turn csv file into list; three columns- Voter ID, County, Candidate
    vote_list = list(csvreader)
    #print(vote_list[0:5])

#Print header
    print("Election Results")
    print("--------------------------------")

#Count total items in list, minus header, equals number of votes
    vote_count = len(vote_list)
    print(f'Total Votes: {vote_count}')
    
    print("--------------------------------")
    
    
    #List number of unique candidate names
    unique_names = []
    for x in vote_list:
        if x[2] not in unique_names:
            unique_names.append(x[2])
    #print(unique_names) #['Khan', 'Correy', 'Li', "O'Tooley"]
    
    #Count votes for each candidate; repeat for each name yielded above
    #better strategy would be to create a while loop to run thru
    #the code block below once for every name in the unique_names list
    Khan_v = 0
    for x in vote_list:
        if x[2] == 'Khan':
            Khan_v = Khan_v + 1
    #print (Khan_v)

    Correy_v = 0
    for x in vote_list:
        if x[2] == 'Correy':
            Correy_v = Correy_v + 1
    #print (Correy_v)   
    
    Li_v = 0
    for x in vote_list:
        if x[2] == 'Li':
            Li_v = Li_v + 1
    #print (Li_v) 
    
    OTool_v = 0
    for x in vote_list:
        if x[2] == "O'Tooley":
            OTool_v = OTool_v + 1
    #print (OTool_v) 
    
    #Percentage of votes each candidate won
    Khan_perc = round((Khan_v / vote_count * 100), 3)
    #for some reason (maybe because the decimals are zero), the round function
    #isn't maintaining the three positions beyond the decimal
    #changing the format with '.3f' maintains the decimals
    Khan_perc = format(Khan_perc,'.3f') 
    #print(Khan_perc)
    Correy_perc = round((Correy_v / vote_count * 100), 3)
    Correy_perc = format(Correy_perc,'.3f') 
    #print(Correy_perc)
    Li_perc = round((Li_v / vote_count * 100), 3)
    Li_perc = format(Li_perc,'.3f') 
    #print(Li_perc)
    OTool_perc = round((OTool_v / vote_count * 100), 3)
    OTool_perc = format(OTool_perc,'.3f') 
    #print(OTool_perc)
    

    
#print statements
    print(f"Khan: {Khan_perc}% ({Khan_v}) ")
    print(f"Correy: {Correy_perc}% ({Correy_v}) ")
    print(f"Li: {Li_perc}% ({Li_v}) ")
    print(f"O'Tooley: {OTool_perc}% ({OTool_v}) ")
    
    print("--------------------------------")
    
#Declare winner
    #Make Dictionary of final candidates
    myDict_perc = {
        "Khan" : Khan_perc,
        "Correy" : Correy_perc,
        "Li" : Li_perc,
        "OTool" : OTool_perc
    }
    #Find the max %
    winner = max(myDict_perc.values())
    #print(f"With {winner}% wins!")
    #print the key that matches the "winner" value
    print("Winner: " + list(myDict_perc.keys())[list(myDict_perc.values()).index(winner)])
    print("--------------------------------")
      
#Write results to a text file:


f = open('PyPollAKH2.txt','w') 

print("Election Results", file = f)
print("--------------------------------", file = f)
print(f'Total Votes: {vote_count}', file = f)
print("--------------------------------", file = f)
print(f"Khan: {Khan_perc}% ({Khan_v}) ", file = f)
print(f"Correy: {Correy_perc}% ({Correy_v}) ", file = f)
print(f"Li: {Li_perc}% ({Li_v}) ", file = f)
print(f"O'Tooley: {OTool_perc}% ({OTool_v}) ", file = f)
print("--------------------------------", file = f)
print("Winner: " + list(myDict_perc.keys())[list(myDict_perc.values()).index(winner)], file = f)
print("--------------------------------", file = f)
            
f.close() 
    
```
